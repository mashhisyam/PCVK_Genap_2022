```python
imgBlack2 = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(imgBlack2)
```

```python
# array([[100, 300],
#        [200, 200],
#        [400, 300],
#        [200, 400]], dtype=int32)

verticies = np.array([[100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
verticies
```

```python
pts = verticies.reshape((-1,1,2)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R, G, B
pts
```
